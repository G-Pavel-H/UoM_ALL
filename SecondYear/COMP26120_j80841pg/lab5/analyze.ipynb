```python
import pandas as pd

data = {'Capacity':[100,200,300,500,700,1000,1200,1800,2500,3000,4000,5000,7000,
8000,10000], 'Time':[0.002555542,0.004181959,0.00631175,0.010670166,0.015339375,0.021682792,0.0258385,0.038785667,
0.054611125,
0.065373042,
0.086822375,
0.105686875,
0.151242416,
0.173193875,
0.214032041]}

df = pd.DataFrame(data)  
```

```python
import matplotlib.pyplot as plt
import numpy as np
a, b = np.polyfit(df['Capacity'], df['Time'], 1)
df.plot(kind='scatter', x = 'Capacity', y = 'Time')
plt.title("Execution time over KP Capacity")
plt.xlabel("Capacity")
plt.ylabel("Time/s")
# plt.ylim(0.02, 0.2)
plt.xlim(0, 6000)
plt.plot(df['Capacity'], a*df['Capacity'] + b)
plt.show()
```
