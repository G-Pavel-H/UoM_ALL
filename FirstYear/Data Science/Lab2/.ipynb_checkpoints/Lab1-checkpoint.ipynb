### Read a CSV file into dataframe


```python
import pandas as pd
import numpy as np
```

```python
olympicsdf = pd.read_csv("olympics.csv",skiprows=1, index_col=0)
```

```python
pd.set_option('display.max_rows',200)
```

```python
olympicsdf.head()
```

```python
print(olympicsdf.columns)
```

```python
olympicsdf = olympicsdf.rename(columns={'01 !':'GS', '02 !' : 'SS', '03 !' : 'BS','01 !.1':'GW', '02 !.1' : 'SW', '03 !.1' : 'BW','01 !.2':'GT', '02 !.2' : 'ST', '03 !.2' : 'BT'})
```
